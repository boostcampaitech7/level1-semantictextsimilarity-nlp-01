```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로와 변수명 리스트
file_paths = [
    "/content/KR-ELCTRA-discriminator-under_sampling0-aug_5.csv",
    "/content/better-english-normalization.csv",
    "/content/lucid.csv",
    "/content/nlp04_nlp04.csv",
    "/content/output_KR_ELECTRA_aug_Data.csv",
    "/content/output_matthewburke_korean_sentiment.csv",
    "/content/redpill-20-epochs.csv",
    "/content/redpill2-20-epochs.csv",
    "/content/sts_klue_roberta-large-epoch7.csv",
    "/content/sts_klue_roberta-large.csv"
]

# 데이터프레임 리스트와 색상 리스트
dataframes = [pd.read_csv(file) for file in file_paths]
labels = [
    'N05: KR-ELCTRA-discriminator-under_sampling0-aug_5',
    'O01: better-english-normalization',
    'N08: lucid',
    'N07: nlp04_nlp04',
    'O07: output_KR_ELECTRA_aug_Data',
    'N01: output_matthewburke_korean_sentiment',
    'N02: redpill-20-epochs',
    'N06: redpill2-20-epochs',
    'N12: sts_klue_roberta-large-epoch7',
    'N10: sts_klue_roberta-large'
]

# 가장 잘나가는 조합
best_combination = ['N01', 'N02', 'N06', 'O01', 'N08']
best_color = 'r'

# 다른 모델들
other_models = [label.split(':')[0] for label in labels if label.split(':')[0] not in best_combination]

# 그래프 그리기
for model in other_models:
    plt.figure(figsize=(20, 14))  # 그래프 크기 조정

    # 가장 잘나가는 조합 그리기
    for short_label in best_combination:
        full_label = next(label for label in labels if label.startswith(short_label))
        idx = labels.index(full_label)
        df = dataframes[idx]
        if 'target' in df.columns:
            df_sorted = df.sort_values(by='target').reset_index(drop=True)
            plt.plot(df_sorted.index, df_sorted['target'], color=best_color, label=f'{full_label} ({best_color})', alpha=0.7, linewidth=2)

    # 추가 모델 그리기
    full_label = next(label for label in labels if label.startswith(model))
    idx = labels.index(full_label)
    df = dataframes[idx]
    if 'target' in df.columns:
        df_sorted = df.sort_values(by='target').reset_index(drop=True)
        plt.plot(df_sorted.index, df_sorted['target'], label=f'{full_label}', alpha=0.7, linewidth=2)

    plt.xlabel('Index')
    plt.ylabel('Target')
    plt.title(f'Comparison of Sorted Target Values with {model}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
